```python
def part_correlation(dataset,threshold):
    col_corr = set()
    corr_matrics = dataset.corr()
    for i in range(len(corr_matrics.columns)):
        for j in range(i):
            if abs(corr_matrics.iloc[i,j]) > threshold:
                column = corr_matrics.columns[i]
                col_corr.add(column)
    return col_corr
                
```

```python
part_correlation(final,0.65)
```

```python
# Hyper Parameter Tuning for choosing best K value
```

```python
error_list = []
accuracy_list = []
for i in range(1,40):
    knnmodel = KNeighborsClassifier(n_neighbors = i)
    score = cross_val_score(knnmodel,X,y,cv = 10)
    error_list.append(1-score.mean())
    accuracy_list.append(score.mean())
```

```python
# Plotting k values against error rate 
```

```python

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_list,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.xlabel('K-value')
plt.ylabel('Error')
plt.title('K-value vs Error rate')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),accuracy_list,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.xlabel('K-value')
plt.ylabel('Accuracy')
plt.title('K-value vs Accuracy rate')
plt.show()
```
